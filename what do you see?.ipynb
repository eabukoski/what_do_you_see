```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline
import seaborn as sns
```

## Data are from the [Early Head Start Research and Evaluation (EHSRE) Study, 1996-2010](https://doi.org/10.3886/ICPSR03804.v5)
#### Enhancing early childhood development for low-income families is the goal of head start programs. The variables selcted below are thought to affect childrens' cognitive and social development. 

```python
# Line Plot
##describe data set
#df.describe
#df.column

early_child= pd.read_csv('EHSRE_data.csv')

plt.figure(figsize=(20, 10))
plt.plot(early_child['POVRATIO'], color='purple')
plt.ylim([0, 400])
plt.ylabel('Values')
plt.title('Income as % of Poverty line for family')
plt.axhline(y=50, linewidth= 3)
plt.show()
```

#### Poverty line levels are used to determine federal services available to families. Poverty levels per family are [calculated](https://poverty.ucdavis.edu/faq/how-poverty-measured-united-states) by multiplying family income by 100 and dividing by the federal poverty level. Unofortunatley, it appears many families in this study earn incomes 50% below their poverty threshold. 
* Values > 100% "above poverty"
* < 125% Values > 100% "near poverty"
* Values < 100% "in poverty"
* Values < 50% "in deep poverty"

```python
#This is the same linegraph using the .plot method from Pandas. 
early_child.plot(y='POVRATIO',kind='line')
```

```python
# Scatter Plot 
#include x and y axis label, explain number of books as bar plot 

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(x=early_child['POVRATIO'], y=early_child['B5P467'], color='purple')
plt.ylim([0, 50])
plt.ylabel('Values')
plt.title('Poverty Level and Number of Books at Home')

plt.subplot(1, 2, 2)
plt.scatter(x=early_child['POVRATIO'], y = early_child['B1P_VCOM'], color='green')
plt.ylabel('Values')
plt.title('Poverty Level and Vocabulary Comprehension')

plt.tight_layout()
plt.show()
```

```python
#Boxplot

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(x = early_child['B5P467'])
plt.ylabel('Number of Books')
plt.title('Number of Books at Home')



plt.subplot(1, 2, 2)
plt.boxplot(x = early_child['B1P_VCOM'])
plt.ylabel('Scores')
plt.title('Vocabulary Comprehension')
plt.show()
```
